## Chocolate Sales Data Analysis(2025)

## Introduction:

This project focuses on analyzing chocolate sales data for the year 2025 to understand key factors influencing revenue and business performance.

## Objective:

To analyze chocolate sales data and identify key insights related to revenue, product performance and sales channel.

```python
import pandas as pd

df = pd.read_csv("chocolate_sales_2025_dataset.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Month'] = df['Date'].dt.month
```

## Monthly Revenue Trend:

```python
import seaborn as sns
import matplotlib.pyplot as plt

monthly_sales = df.groupby('Month')['Revenue_USD'].sum()
sns.lineplot(x = monthly_sales.index, y = monthly_sales.values)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

The monthly trend shows variations in revenue across different months, indicating possible seasonal patterns in chocolate series.

## Research Question:

Which factors(product type, brand, and sales channel) influence revenue the most in chocolate sales?

## Revenue by Product Type:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x = "Product_Type", y = "Revenue_USD", data = df)
plt.title("Revenue by Product Type")
plt.xticks(rotation=45)
plt.show()
```

#The analysis shows that Chocolate Box contributes the highest revenue, followed by White Chocolate, whereas Dark Chocolate generates the least revenue. This suggests that customers demand varries accross product types, making product type as important factor affecting overall revenue. 

## Revenue by Brand:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x = "Brand", y="Revenue_USD", data = df)
plt.title("Revenue by Brand")
plt.xticks(rotation=45)
plt.show()
```

This analysis shows that Mars is the top performing brand in terms of revenue,followed by Toblerone and Nestle. In contrast, brands such as Cadbury, Lindt, Hershey and Ferrero generate relatively lower revenue. This highlights that brand preference significantly influences customer purchasing behavior and overall revenue generation.

## Revenue by Sales Channel:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot( x = "Sales_Channel", y = "Revenue_USD", data = df)
plt.title("Revenue by Sales Channel")
plt.xticks(rotation=45)
plt.show()
```

This analysis shows that Supermarket contribute the highest revenue among all sales channels. However, other channels such as online,  Convenience Store and Specialty Store generate similar revenue levels. This suggests that while sales channel has some influence, it is not the most dominant factor affecting overall revenue.

## Units Sold vs Revenue:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot( x = "Units_Sold", y = "Revenue_USD", data =df)
plt.title("Units Sold vs Revenue")
plt.xticks(rotation=45)
plt.show()
```

The scatter plot shows a clear positive correlation between units sold and revenue.As the number of units sold increases,revenue also increases significantly.This suggests that sales volumne is a key driver of revenue in chocolate sales. 

## Correlation Matrix:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(numeric_only = True), annot = True)
plt.title("Correlation Matrix")
plt.xticks(rotation = 45)
plt.show()
```

The correlation analysis reveals that revenue is strongly influenced by units sold (correlation of 0.7) and moderately influenced by price (correlation of 0.6). This suggests that while pricing contributes to revenue, the volumne of sales is the most significant driver of overall revenue generation.

## Conclusion:

Based on the analysis, revenue in chocolate sales is influenced by multiple factors including product type, brand, sales channel, units sold and price.
Among these, units sold has the strongest impact on revenue, followed by price.Product type and brand also significantly affect performance, with certain categories and brands generating higher revenue.While sales channel contributes to revenue differences, its impact is relatively smaller.Overall, the analysis shows that increasing sales volumne, along with strategic pricing and focusing on high-performance products and brands, can maximize revenue.
